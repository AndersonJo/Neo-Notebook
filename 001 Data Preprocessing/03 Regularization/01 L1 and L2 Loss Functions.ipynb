# L1 and L2 Loss Functions

#### L1-norm loss function (Least absolute deviations - LAD)

$$ s = \sum^n_{i=1} | y_i - f(x_i) | $$

#### L2-norm loss function (Least sqaures error - LSE)

$$ s = \sum^n_{i=1} \left( y_i - f(x_i) \right)^2 $$


L1-norm loss function의 경우 outlier에 강합니다. <br>
L2의 경우 loss값에 square를 하기 때문에 outlier의 그 값이 더욱 크게 되고 다른 일반적인 값들에 맞춰서 loss를 minimize하는게 아니라 outlier에 맞춰서 minimize가 될 수 있습니다.
