### Data Loading

NIV 영어 성경에서 'money'가 들어간 단어가 들어간 구절들만 모두 뽑은 텍스트 파일을 불러 옵니다.

```python
with codecs.open('../../data/simple-research/bible-money.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
```

### Data Cleasing

쓸모없는 단어들은 삭제합니다.<br>
1. **ADP** 는 adposition으로서 on, of, at, with, by, into, under같은 것들입니다.
2. **CONJ** 는 Conjunction으로서 and, or, but, if, while, although같은 것들입니다.
3. **DET** 는 Determiner, article로서 the, a, some, most, every, no, which같은 것들입니다.
4. **PRT** 는 Particle로서 at, on, out, over, per, that, up, with같은 단어들입니다.
5. **IN** 는 with, if, for, of, among, like 같은 단어들입니다.
6. **CC** 는 and, but, or같은 단어들입니다.
7. **PRP** 는 they, themselves, he, it, they, it, he, him 같은 단어들을 뜻합니다.
8. **VBZ** 는 is를 가르킵니다.
9. **VBD** 는 was 를 가르킵니다.
6, **TO** 는 to 단어를 뜻합니다.
7. **.**는 Punctuation Marks로서 .,;! 같은 것들입니다.
8. **:**는 :;같은 단어들입니다.


```python
tokenized_text = nltk.word_tokenize(text)
tagged = nltk.pos_tag(tokenized_text)
print filter(lambda t: t[0] in ['was'], tagged)
filters = ['DT', 'CONJ', 'DET', 'PRT', 'CC', 'TO', 'IN', 'PRP', 'VBZ', 'VBD', '.', ':', ',']
tagged = filter(lambda t: t[1] not in filters, tagged)
tokenized_text = map(lambda t: t[0], tagged)
print tagged
```

```python
freq = nltk.FreqDist(tokenized_text)
freq.most_common()
```

```python
regex = re.compile(r'^(NN|VB).*')
tagged2 = filter(lambda t: regex.match(t[1]), tagged)
tokized2 = map(lambda t: t[0], tagged2)
nltk.FreqDist(tokized2).most_common()

```
