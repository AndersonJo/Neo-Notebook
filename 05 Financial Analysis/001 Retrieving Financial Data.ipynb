## Code

```python
%pylab inline
import pandas as pd
import pandas_datareader as pdr
import numpy as np

from googlefinance import getQuotes
from datetime import datetime

start = datetime(2015, 1, 1)
end = datetime(2017, 5, 31)
```

## Google Finance

```python
data = pdr.data.DataReader('GOOGL', 'google', start, end)
```

```python
data.plot(y=['Open', 'High', 'Low', 'Close'], grid=True)
```
