# Normal Distribution

일반적으로 수학자들은 Normal distribution이라고 부르고, 물리학자들의 경우 Gaussian distribution 이라고 하며, <br>
사회 과학자들은 Bell Curve 라고도 부릅니다. 

Standard normal distribution은 $ \mu = 0 $ 그리고 $ \sigma^2 = 1 $ 값을 줍니다.


### $$ P(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{ - \frac{(x-\mu)^2}{2 \sigma^2} } $$

```python
%pylab inline
import numpy as np
import scipy.stats
```

### Normal Distribution with Numpy

```python
data = np.random.normal(size=30000)

title('Normal Distribution with Numpy')
grid()
hist(data, bins=100, normed=True)

print
```


$$p(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }} e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} } $$

```python
def normal_distribution(mu=0, std=1):
    bins = np.arange(0, 1000, 1)
    return 1/np.sqrt(2*np.pi * std) * np.exp(- (bins - bins.mean())**2/(2 * std))

data = normal_distribution()

hist(data, bins=100)
print 
```
