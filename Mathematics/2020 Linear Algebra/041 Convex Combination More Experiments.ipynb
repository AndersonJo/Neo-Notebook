# Convex Combination

vector space안의 any points $ x_1, x_2, ... x_k $가 다음의 형태를 취할때 convex combination입니다.<br>
이때 $ \theta_i \ge 0 $ 이며, $ \theta_1 + \theta_2 + ... + \theta_k = 1 $ 입니다.

$$ x = \theta_1 x_1 + \theta_2 x_2 + ... + \theta_k x_k $$ 

### Example 

예를 들어서 points $ P_0, P_1 $ 이 있고, 2점 사이의 any point $ P $는 다음과 같이 나타낼수 있습니다. 

$$ P = \theta P_0 + (1-\theta)P_1 $$

위의 공식은 다음과 같습니다. 

$$ P = \alpha_0 P_0 + \alpha_1 P_1$$

즉 affine combination이며 이때 $ \alpha_0 + \alpha_1 = 1 $ 이라면 convex combination 입니다. 

### Example 1

다음과 같이 Affine Vector 이면서, Convex Combination ($ \sum_i \alpha_i = 1 $)인 2개의 point를 convex combination으로 plotting할때 다음과 같은 결과가 나옵니다. 

$$ \begin{align}
x & = \theta \cdot \begin{bmatrix}1 \\ 3 \end{bmatrix} + (1-\theta) \cdot \begin{bmatrix} 3 \\ 4 \end{bmatrix}  \\
x &= \alpha_1 \cdot \begin{bmatrix}1 \\ 3 \end{bmatrix} + \alpha_2 \cdot \begin{bmatrix} 3 \\ 4 \end{bmatrix} \\
\end{align}
$$

즉 Affien 공식 $ \theta x_1 + (1-\theta)x_2 , \ \theta \in \textbf{R} $과 일치하기 때문에 한 선안에, 점이 찍히는게 보입니다.
또한 theta $ \theta $값이 커지면서 magnitude가 달라지게 됩니다. 

```python
def convex_combination(a, b, theta=0.5):
    return theta*a + theta*b

def visualize(o, a, b):
    plot(*np.dstack([o, a])[0], color="#333333")
    plot(*np.dstack([o, b])[0], color="#333333")

    for theta in np.arange(0, 1, 0.1):
        c = convex_combination(a, b, theta)
        print('theta: %.1f \tconvex combination: %s'% (theta, c))
        scatter(*np.dstack([o, c])[0])
        
        
o = np.array([0, 0])
a = np.array([1, 3])
b = np.array([3, 4])        
visualize(o, a, b)
```

### Example 2

```python
o = np.array([0, 0])
a = np.array([-1, 10])
b = np.array([5, -30])
visualize(o, a, b)
```
