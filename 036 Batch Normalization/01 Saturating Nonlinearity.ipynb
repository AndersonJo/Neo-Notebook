# Saturating Nonlinearity

### Saturate 사전적 정의

 * put (a device) into a state in which no further increase in current is achievable.
 * fill (something or someone) with something until no more can be held or absorbed.
 
즉 좌표평면에서 이야기를 하자면, y값에 한정된 범위 (0에서 1사이 같은..)가 있다면 saturating 이고, 0에서 무한대 (relu 같은것) 은 non-saturating 입니다.<br>
즉 더이상 증가시킬수 없는 상태로 만들수 있으면 saturating 입니다.

### Non-Saturating 정의

다음의 조건을 만족할때 함수 $ f $는 non-saturating 입니다.

$$ \left| \lim_{z \rightarrow -\infty} f(z) \right| = +\infty \ \lor \ \left| \lim_{z \rightarrow +\infty} f(z) \right| = +\infty $$

> 여기서 $ \lor $ 의 의미는 logical disjunction 으로서 **or**을 나타냅니다.<br>
> 즉 2개의 statements중에서 하나만 True이면 True이고, 2개다 틀릴시 False라는 뜻


### Saturating 정의 

Non-Saturating이 아닌것

## ReLU -> Non-Saturating 

ReLU $ f(x) = max(0, x) $ 는 Non-Saturating 입니다.

$$ \lim_{z \rightarrow +\infty} f(z) = + \infty $$

```python
%pylab inline
import numpy as np

def relu(x):
    return np.maximum(x, 0)

x = np.arange(-10, 10, 0.1)
y = relu(x)

title('ReLU - Non-Saturating')
plot(x, y)
grid()
```

## Sigmoid - Saturating

sigmoid $ \frac{1}{1+e^{-z}} $ 는 saturating입니다. 왜냐하면 [0, 1] 이기 때문입니다.

```python
def sigmoid(x):
    return 1/(1+np.e**(-x))

x = np.arange(-10, 10, 0.1)
y = sigmoid(x)

plot(x, y)
grid()
```
