```python
%pylab inline
import numpy as np
import cv2
```

```python
img1 = cv2.imread('../data/images/baseball.jpg', cv2.CV_LOAD_IMAGE_COLOR)
img2 = cv2.imread('../data/images/girl.jpg', cv2.CV_LOAD_IMAGE_COLOR)
```

```python
print 'Shape:',  img1.shape
print 'Type :', img1.dtype
print 'Total number of pixels:', img1.size
```

```python
pylab.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
pylab.show()
pylab.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
pylab.show()
```

```python
b1, g1, r1 = cv2.split(img1)
b2, g2, r2 = cv2.split(img2)
img3 = cv2.merge((b1, g1, r2))

pylab.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
pylab.show()
```
