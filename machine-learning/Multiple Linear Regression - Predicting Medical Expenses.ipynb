```python
%pylab inline
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from pandas.stats.api import ols
```

```python
data = pd.read_csv('../data/multiple-linear-regression/insurance.csv')
data['smoker'] = data['smoker'].apply({'yes': 1, 'no': 0}.get)
data['sex'] = data['sex'].apply({'male': 1, 'female': 2}.get)
data = data[['age', 'sex', 'bmi', 'children', 'smoker', 'region', 'expenses']]
```

```python
print data.corr()
```

```python
figures = scatter_matrix(data, figsize=(10, 10))
```
