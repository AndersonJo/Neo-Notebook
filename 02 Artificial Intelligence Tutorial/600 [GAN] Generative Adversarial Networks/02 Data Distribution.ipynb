### Binomial Distribution

Binomial distribution를 갖고 있는 random variable $ X $는 yes/no trials에서 성공확률 $ P $를 나타냅니다. 

<span style="color:red"> TODO 일단 아래는 다시 써야 함</span>

$$ P(E)   = {n \choose k} p^k (1-p)^{ n-k} $$

> **WikiPedia 설명** <br>
> In general, if the random variable X follows the binomial distribution with parameters n ∈ ℕ and p ∈ [0,1], we write X ~ B(n, p). The probability of getting exactly k successes in n trials is given by the probability mass function:


* $ E(X) $ 는 expected value로서 distribution의 평균(mean)을 의미합니다.

```python
n = 10
p = 0.3
k = np.arange(0, 21)

binomial = stats.binom.pmf(k, n, p)

plot(k, binomial, 'o-')
title(f'Binomial: n={n}, p={p:.2}', fontsize=15)
xlabel('Number of Successes')
ylabel('Probability of Success')
```

### Poisson Distribution

특정 시간동안 $ \lambda $ rate 값에 따라, 얼마나 많이 event가 일어날지를 나타냅니다.

<span style="color:red"> TODO 공식 추가</span>

```python
rate = 2
n = np.arange(0, 10)
poisson = stats.poisson.pmf(n, rate)

plot( poisson)
title(f'Poisson: $\lambda $={rate}')
xlabel('Number of accidents')
ylabel('Probability of number of accidents')
```
