```python
%pylab inline
import numpy as np
import pandas as pd
```

```python
def logistic(z): 
    return 1 / (1 + np.exp(-z))

def logit(p):
    return np.log(p/(1-p))

def odds(p):
    return p/(1-p)
```

```python
p = np.arange(0.01, 1, 0.05)
odds_data = p/(1-p)
logit_data = np.log(odds_data)

ylim(-5, 5)
plot(p, odds_data, label='odds')
plot(p, logit_data, label='logit')
legend(bbox_to_anchor=(0.25, 1))
xlabel('Probability')
grid()

```
